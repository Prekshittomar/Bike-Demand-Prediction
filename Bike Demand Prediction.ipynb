```python
bt['demand']=np.log(bt['demand'])
print(bt)
```

```python
# Autocorrelation in the demand column
t_1=bt['demand'].shift(+1).to_frame()
t_1.columns=['t-1']

t_2=bt['demand'].shift(+2).to_frame()
t_2.columns=['t-2']

t_3=bt['demand'].shift(+3).to_frame()
t_3.columns=['t-3']

bt_lag=pd.concat([bt,t_1,t_2,t_3],axis=1)
print(bt_lag)
```

```python
bt_lag=bt_lag.dropna()
```

```python
print(bt_lag)
```

```python
# create dummy variables and drop first to avoid dummy variable using get_dummies
bt_lag.dtypes
```

```python
bt_lag['season']=bt_lag['season'].astype('category')
bt_lag['month']=bt_lag['month'].astype('category')
bt_lag['hour']=bt_lag['hour'].astype('category')
bt_lag['holiday']=bt_lag['holiday'].astype('category')
bt_lag['weather']=bt_lag['weather'].astype('category')

bt_lag=pd.get_dummies(bt_lag,drop_first=True)
```

```python
bt_lag.head()
```

```python
# Split the X and Y in the training and testing set

# demand is time dependent or the time series

Y=bt_lag[['demand']]
X=bt_lag.drop(['demand'],axis=1)
```

```python
X.head()
```

```python
# create the size for 70% of the data
tr_size=0.7*len(X)
print(tr_size)
```

```python
tr_size=int(tr_size)
print(tr_size)
```

```python
# Create the train and test using tr_size
x_train=X.values[0:tr_size]
x_test=X.values[tr_size:len(X)]

y_train=Y.values[0:tr_size]
y_test=Y.values[tr_size:len(Y)]
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
std_reg=LinearRegression()
std_reg.fit(x_train,y_train)
```

```python
r2_train=std_reg.score(x_train,y_train)
r2_test=std_reg.score(x_test,y_test)
```

```python
Y_predict=std_reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse=math.sqrt(mean_squared_error(y_test,Y_predict))
```

```python
print(r2_train)
print(r2_test)
print(rmse)
```

```python
# Calculate the RMSLE (root mean square logaritm error) and compare result
y_test_e=[]
y_predict_e=[]

for i in range(0,len(y_test)):
    y_test_e.append(math.exp(y_test[i]))
    y_predict_e.append(math.exp(Y_predict[i]))
    
```

```python
# Do the sum of the logs and squares
log_sq_sum=0.0

for i in range(0,len(y_test)):
    log_a=math.log(y_test_e[i]+1)
    log_p=math.log(y_predict_e[i]+1)
    log_diff=(log_p-log_a)**2
    log_sq_sum=log_sq_sum+log_diff
    
rmsle=math.sqrt(log_sq_sum/len(y_test))
print(rmsle)
```
